```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def julia(z,c):
    """ Complex Function"""
    oparaterd_point = z*z + c 
    return(oparaterd_point)
```

```python
c = complex(-0.4,0.6)

matris0 = np.zeros((500,500))
points_range = np.linspace(-2,2,500)
```

```python

complexpart = 0
realpart = 0

for complx in points_range:
    for real in points_range:
        z = complex(real,complx)
        for i in range(1000):
            z = julia(z,c)
            if abs(z) > max(abs(c), 2):# max(abs(c), 2) come from https://stackoverflow.com/questions/27953051/how-to-improve-poor-visuals-of-julia-sets-made-using-matplotlib
                matris0[complexpart, realpart]=i
                break
        realpart += 1
    complexpart += 1
    realpart = 0
    
```

```python
plt.axis('off')
plt.title('c =-0.4+0.6i ,Check 1000 time',fontsize=15)
plt.imshow(matris0, cmap="viridis", extent=[-2,2,-2,2])
```
