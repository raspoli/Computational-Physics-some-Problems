```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.patches import Polygon 
```

## Functions

```python
def middle_point(p1, p2):
    """ finde poit that is between p1 and p2"""
    return[(p1[0] + p2[0])/2,(p1[1] + p2[1])/2]
```

```python
def draw(level, p1, p2, p3):
    ''' draw Sierpinski Triangle in specified level'''
    if level == 1:
        points.append([p1, p2, p3])
    else:
        p4 = midpoint(p1, p2);
        p5 = midpoint(p2, p3);
        p6 = midpoint(p1, p3);
        draw(level - 1, p1, p4, p6);
        draw(level - 1, p4, p2, p5);
        draw(level - 1, p6, p5, p3);
```

## Process

```python
points = []
level = 6
draw(level,[0,0],[1,0],[0.5,np.sqrt(3)/2])
points = np.array(points)
```

## Visualize
