```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

## Functions

```python
def F1(vector): # First Mapping function
    Rotating = np.array([[0,0],[0,0.16]])
    bias = np.array([0,0])
    return(np.dot(Rotating,vector) + bias)
```

```python
def F2(vector): # Second Mapping function
    Rotating = np.array([[0.85,0.04],[-0.04,0.85]])
    bias = np.array([0,1.60])
    return(np.dot(Rotating,vector) + bias)
```

```python
def F3(vector): # Third Mapping function
    Rotating = np.array([[0.20,-0.26],[0.23,0.22]])
    bias = np.array([0,1.60])
    return(np.dot(Rotating,vector) + bias)
```

```python
def F4(vector): # forth Mapping function
    Rotating = np.array([[-0.15,0.28],[0.26,0.24]])
    bias = np.array([0,0.44])
    return(np.dot(Rotating,vector) + bias)
```

## Process

```python
points = np.array([[np.random.random(),np.random.random()]])
point_number = 30000
Functions = [F1, F2, F3, F4]
current = 0
for _ in range(point_number):
    c = np.random.choice([0,1,2,3],p=[0.01,0.85,0.07,0.07])
    points = np.append(points,Functions[c](points[current]).reshape(1,2),axis=0)
    current += 1 
    
points = np.transpose(points)
```

```python
fig = plt.figure(figsize=(5,5))
plt.axis('off')
plt.scatter(points[0], points[1], s = 0.2, edgecolor ='green') 
```
