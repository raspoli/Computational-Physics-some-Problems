```python
import numpy as np
import matplotlib.pyplot as plt
```

## Functions

```python
def Rotating_Matrix(theta,vector):
    """Rotate Direction vector"""
    r = np.array(( (np.cos(theta), -np.sin(theta)),
                  (np.sin(theta),  np.cos(theta)) ))
    rotated_vector = r.dot(vector)
    
    return(rotated_vector)
```

## Process

```python
level = [3,7,12,17]
p = [[],[],[],[]]
p_p = [[],[],[],[]]
```

```python
for k,l in enumerate(level):
    points = np.array([[-1,0],[1,0]])

    for _ in range(l):
        points_number = points.shape[0]
        new_points = np.zeros(((2*points_number-1),2))
    
        for i in range(points_number-1):
       
            new_points[2*i] = points[i]
        
            vector = points[i+1] - points[i]
            r_vector =  Rotating_Matrix(((-1)**i) * np.pi/4,np.sqrt(2)*vector)
           
            new_points[2*i+1] = points[i] + r_vector/2
       
        new_points[2*points_number-2] = points[points_number-1]  
        points = new_points
        
    point_prime = points[range(int(len(points)/2)+1)]
    p[k] = np.transpose(points)
    p_p[k] = np.transpose(np.array(point_prime))
```

## Visualize

```python
fig, axes = plt.subplots(2, 2,figsize = (16,16))

axes[0,0].axis('off')
axes[0,1].axis('off')
axes[1,0].axis('off')
axes[1,1].axis('off')

axes[0,0].set_title('level %i'%level[0],fontsize=30)
axes[0,1].set_title('level %i'%level[1],fontsize=30)
axes[1,0].set_title('level %i'%level[2],fontsize=30)
axes[1,1].set_title('level %i'%level[3],fontsize=30)

axes[0,0].plot(p[0][0],p[0][1],'m')    
axes[0,0].plot(p_p[0][0],p_p[0][1],'b')

axes[0,1].plot(p[1][0],p[1][1],'m')    
axes[0,1].plot(p_p[1][0],p_p[1][1],'b')

axes[1,0].plot(p[2][0],p[2][1],'m')    
axes[1,0].plot(p_p[2][0],p_p[2][1],'b')

axes[1,1].plot(p[3][0],p[3][1],'m')    
axes[1,1].plot(p_p[3][0],p_p[3][1],'b')
```
