```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lattice_size = 100
p = 0.55
```

```python
lattice = np.random.choice([0,1],p=[1-p,p],size=((lattice_size,lattice_size)))
```

```python
fig = plt.figure(figsize=(8,8))
plt.title('P =%.2f , size=%i '%(p,lattice_size) , fontsize = 15)
plt.axis('off')
plt.imshow(lattice,'binary')
#plt.savefig('Percolation fig4')
```
